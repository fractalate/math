# Interactive Charts

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from ipywidgets import interactive
import ipywidgets as widgets

%matplotlib widget

GRID_MIN = -6
GRID_MAX = 6

def apply_line_style(line, vector):
    line.axes.annotate(
        '',
        xytext=(0, 0),
        xy=(vector[0], vector[1]),
        arrowprops=dict(color=line.get_color(), arrowstyle="simple"),
        size=10
    )

def configure_graph_grid():
    GRID_MIN = -6
    GRID_MAX = 6

    plt.axis('square')
    plt.axis((GRID_MIN, GRID_MAX, GRID_MIN, GRID_MAX))

    plt.plot([GRID_MIN, GRID_MAX], [0, 0], 'black')
    plt.plot([0, 0], [GRID_MIN, GRID_MAX], 'black')
    plt.grid()

def do_update(v1, v2):
    fig.clf()
    configure_graph_grid()
    line1 = plt.plot([0, v1[0]], [0, v1[1]], label='Vector 1')[0]
    apply_line_style(line1, v1)
    line2 = plt.plot([0, v2[0]], [0, v2[1]], label='Vector 2')[0]
    apply_line_style(line2, v2)
    plt.legend()
    plt.title('Interactive Chart')
    plt.draw()

def handle_event(v1_x, v1_y, v2_x, v2_y):
    do_update([v1_x, v1_y], [v2_x, v2_y])

def slider_config(value):
    return widgets.IntSlider(min=GRID_MIN, max=GRID_MAX, step=1, value=value)

vector_1 = [3, 4]
vector_2 = [1, 3]

plt.close()
fig = plt.figure()
slider_controls = interactive(
    handle_event,
    v1_x=slider_config(1),
    v1_y=slider_config(2),
    v2_x=slider_config(3),
    v2_y=slider_config(4),
)
display(slider_controls)
```
